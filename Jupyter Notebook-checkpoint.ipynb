```python
students_perfomance = students_perfomance.rename(columns = {'parental level of education': 'parental_level_of_education',
'test preparation course': 'test_preparation_course', 'math score': 'math_score', 'reading score' : 'reading_score',
'writing score': 'writing_score'})
```

```python
students_perfomance.query("writing_score > 74") #выводит строки с наложенным на них в скобках условием
```

```python
students_perfomance.query("gender == 'female'& writing_score > 70") #female надо тоже заключать в кавычки, т.к. это строка, это более короткий способ вывода строк, соответствующих женщинам с баллом > 70
```

```python
students_perfomance.filter(like = "score", axis=1) #считывает столбцы (axis=1, axis=0 - по строкам), которые содержат (like) слово score
```

```python
students_perfomance.groupby('gender').mean() #выводит среднее значение всех величин, которые относятся к группе gender (сначала происходит группировка по признаку gender методом groupby, а потом взятие среднего от всех численных величин)
```
